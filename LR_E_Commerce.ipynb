### Predictions 
it returns an array (list) of predicted values, which you can add as a new column in your dataset if needed.

```python
predictions = lm.predict(X_test)
predictions #--> It predicts the target (Y) values using the trained model (lm) and the test data (X_test).
```

#### now, we are going to plot a scatter plot using this predctions against the actual values present in the dataset 

```python
sns.scatterplot(x=predictions, y=y_test)
plt.xlabel("Predictions")
plt.title("Eveluation of our LM model")
plt.show()
```

### Eveuation of the matrices
##### 1) Check R Score --> It just means that the model fits the training data well, but that doesnt guarantee good predictions on new data.
###### if R is colse to 1 --> model explains most of the varicence (Good)
###### If R is colse to 0 --> Model doesnt explain much variance (Bad)

```python
from sklearn.metrics import r2_score
r2 = r2_score(y_test, predictions)
print(f"R Score{r2}")
```

Our R Score0.9808757641412774 value is close to 1 --> It just means that the model fits the training data well, but that doesnt guarantee good predictions on new data.

##### 2) RMSE (Root Mean Squared Error) , MSE (Mean Squared Error) And MAR (Mean Absolute Error)
##### --> They tell us how far off our predictions are from the actual values. 
##### --> It shows how much error the model makes

```python
from sklearn.metrics import mean_squared_error , mean_absolute_error
rmse = np.sqrt(mean_squared_error(y_test, predictions))
mse = mean_squared_error(y_test, predictions)
mae = mean_absolute_error(y_test, predictions)
print(f"RMSE = {rmse}")
print(f"MSE = {mse}")
print(f"MAE = {mae}")
```

# Final Conclusion 

It can be tricky to interpret the information in this analysis. According to the model, the most significant factor for clients is not the time spent on the app or website, but their length of membership. In other words, according to the data, the amount of time that the customer spends on the desktop website has almost nothing to do with the amount of money they will spend.

Customers spend significantly more time on the desktop website compared to the desktop application. However, this does not lead to more purchases, suggesting that users may be facing difficulties in navigation or usability issues on the website. This indicates a possible need for improvements in the user experience to enhance engagement and increase conversion rates.


We could interpret this in two different ways. Firstly, this could mean that the desktop website needs more work to make its visitors buy more. Secondly, it could mean that people tend to be more influenced by mobile applications of online stores than by desktop websites.
